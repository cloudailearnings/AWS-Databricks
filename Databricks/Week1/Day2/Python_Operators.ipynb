# 🧮 Python Operators 

Business & real-world examples for **all operator types**

## 1️⃣ Arithmetic Operators (Salary & Finance)

```python
basic_salary = 50000
hra = 10000
bonus = 5000

total_salary = basic_salary + hra + bonus
monthly_salary = total_salary / 12
tax = total_salary * 0.10
remaining = total_salary - tax

total_salary, monthly_salary, tax, remaining
total_salary
```

## 2️⃣ Assignment Operators (Incremental Updates)

```python
salary = 50000
salary += 5000   # increment
#salary -= 2000   # deduction

salary *= 2      # appraisal
salary /= 2      # normalization


salary
```

## 3️⃣ Comparison Operators (Eligibility Check)

```python
experience = 6
salary = 70000

experience >= 5, salary > 60000, salary == 70000, salary != 50000
```

## 4️⃣ Logical Operators (Business Rules)

```python
is_employee_active = True
experience = 6
salary = 70000

eligible_bonus = experience >= 5 and salary > 60000
eligible_promotion = experience >= 8 or salary > 90000
not_active = not is_employee_active

eligible_bonus, eligible_promotion, not_active
```

## 5️⃣ Identity Operators (Object Comparison)

```python
emp1 = ['Rajesh', 101]
emp2 = ['Rajesh', 101]
emp3 = emp1

emp1 is emp2, emp1 == emp2, emp1 is emp3
```

## 6️⃣ Membership Operators (Access Control)

```python
allowed_users = ['admin', 'manager', 'hr']
user_role = 'manager'

user_role in allowed_users, 'abc' not in allowed_users
```

## 7️⃣ Bitwise Operators (Low-Level / Flags Example)

```python
READ = 4
WRITE = 2
EXECUTE = 1

user_permission = READ | WRITE

can_read = user_permission & READ
can_execute = user_permission & EXECUTE

can_read, can_execute
```

## 8️⃣ Ternary Operator (Quick Decision Making)

```python
salary = 75000
grade = 'Senior' if salary >= 70000 else 'Junior'

grade
```

##  Payroll Eligibility CHECK

```python
def payroll_eligibility(salary, experience, active):
    bonus = salary * 0.10 if salary > 60000 and experience >= 5 else 0
    promotion = experience >= 8 or salary >= 90000
    return bonus, promotion

payroll_eligibility(75000, 6, True)
```
