# INPUT AND OUTPUT

```python
name = input("Enter employee name: ")
experience = int(input("Enter years of experience: "))

if experience >= 5:
    print(f"{name} is a Senior Employee")
else:
    print(f"{name} is a Junior Employee")
```

```python
tax_rate = 0.30

def calculate_tax(salary):
    tax_rate = 0.20
    return salary * tax_rate

print("Tax inside function:", calculate_tax(50000))
print("Tax rate outside function:", tax_rate)
```

```python
def employee_summary():
    name = input("Name: ")
    salary = float(input("Monthly Salary: "))
    
    yearly_salary = salary * 12
    bonus = yearly_salary * 0.10
    
    print("\n--- Employee Summary ---")
    print("Name:", name)
    print("Yearly Salary:", yearly_salary)
    print("Bonus:", bonus)

employee_summary()
```
