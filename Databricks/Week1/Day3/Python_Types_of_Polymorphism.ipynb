# Python – Types of Polymorphism

## 1. Method Overriding (Runtime Polymorphism)

**Definition:** Child class provides its own implementation of a parent method.

```
Order
  |
PremiumOrder
```

````python
class Order:
    def get_type(self):
        return 'Regular Order'

class PremiumOrder(Order):
    def get_type(self):
        return 'Premium Order'

print(Order().get_type())
print(PremiumOrder().get_type())
````

## 2. Method Overloading (Using Default Arguments)

**Note:** Python does not support traditional method overloading.
It is achieved using default arguments or *args.
```
calculate(price)
calculate(price, discount)
```

````python
class PriceCalculator:
    def calculate(self, price, discount=0):
        return price - (price * discount)

pc = PriceCalculator()
print(pc.calculate(1000))
print(pc.calculate(1000, 0.10))
````

## 4. Duck Typing

**Definition:** Object behavior matters, not object type.

```
pay()
 |  \
Cash Card
```

````python
class CashPayment:
    def pay(self):
        return 'Cash Payment'

class CardPayment:
    def pay(self):
        return 'Card Payment'

def make_payment(payment):
    print(payment.pay())

make_payment(CashPayment())
make_payment(CardPayment())
````

##  Summary

- Method Overriding → Runtime polymorphism
- Method Overloading → Default args / *args
- Duck Typing → Behavior-based polymorphism

**Polymorphism = One interface, many behaviors**
