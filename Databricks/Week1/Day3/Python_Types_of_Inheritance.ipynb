# Python – Types of Inheritance

## 1. Single Inheritance

**Definition:** One child class inherits from one parent class.

### 
```
Order
  |
OnlineOrder
```

````python
class Order:
    def order_type(self):
        return 'Regular Order'

class OnlineOrder(Order):
    def payment_mode(self):
        return 'Online Payment'

o = OnlineOrder()
print(o.order_type())
print(o.payment_mode())
````

## 2. Multiple Inheritance

**Definition:** One child class inherits from multiple parent classes.

```
OnlinePayment   Discount
      \           /
       \         /
        OnlineOrder
```

````python
class OnlinePayment:
    def payment(self):
        return 'Online Payment'

class Discount:
    def discount(self):
        return '10% Discount'

class OnlineOrder(OnlinePayment, Discount):
    pass

o = OnlineOrder()
print(o.payment())
print(o.discount())
````

## 3. Multilevel Inheritance

**Definition:** Inheritance chain (Parent → Child → Grandchild).

```
Order
  |
PremiumOrder
  |
InternationalOrder
```

````python
class Order:
    def order(self):
        return 'Order'

class PremiumOrder(Order):
    def priority(self):
        return 'High Priority'

class InternationalOrder(PremiumOrder):
    def shipping(self):
        return 'International Shipping'

o = InternationalOrder()
print(o.order())
print(o.priority())
print(o.shipping())
````

## 4. Hierarchical Inheritance

**Definition:** Multiple child classes inherit from the same parent.

```
        Order
      /   |    \
Online  Store  Phone
```

````python
class Order:
    def order_type(self):
        return 'Base Order'

class OnlineOrder(Order):
    pass

class StoreOrder(Order):
    pass

class PhoneOrder(Order):
    pass

print(OnlineOrder().order_type())
print(StoreOrder().order_type())
print(PhoneOrder().order_type())
````

## 5. Hybrid Inheritance

**Definition:** Combination of two or more types of inheritance.
```
        Order
          |
    OnlinePayment
       /      \
 Discount   PremiumOnlineOrder
```

````python
class Order:
    def order(self):
        return 'Order'

class OnlinePayment(Order):
    def payment(self):
        return 'Online Payment'

class Discount:
    def discount(self):
        return '10% Discount'

class PremiumOnlineOrder(OnlinePayment, Discount):
    pass

o = PremiumOnlineOrder()
print(o.order())
print(o.payment())
print(o.discount())
````

## ✅ Summary

- Single: One parent → one child
- Multiple: Many parents → one child
- Multilevel: Chain inheritance
- Hierarchical: One parent → many children
- Hybrid: Combination
