# 🐍 Python – Functions & Arrays


## 1️⃣ Functions – Reusable Business Logic

Used in

ETL transformations

Logical builds

Validation rules


```python
def calculate_final_price(price, discount, gst):
    discounted_price = price - (price * discount)
    final_price = discounted_price * (1 + gst)
    return final_price

final_amount = calculate_final_price(12000, 0.10, 0.18)
final_amount
```

### Function – Order Classification

```python
def classify_order(amount):
    if amount > 100000:
        return "Premium Order"
    else:
        return "Regular Order"

classify_order(150000)
```

## 2️⃣ Lambda Functions -  small, one-line function, often used with filters & transformations.

## Used in

## Data filtering

## Inline transformations

## Pandas / Spark operations

```python
discount_price = lambda price: price * 0.90
discount_price(5000)
```

### Lambda with Filter – High Value Orders

```python
orders = [20000, 80000, 150000]

high_value_orders = list(filter(lambda x: x > 110000, orders))
high_value_orders
```

## 3️⃣ Arrays (Lists) Python supports arrays using lists or the array module

```python
order_amounts = [25000, 60000, 120000, 45000]
order_amounts
```

### Array Processing Using Function

```python
def apply_gst(amounts):
    result = []
    for amt in amounts:
        result.append(amt * 2)
    return result

apply_gst(order_amounts)
```

### Array Processing Using Lambda

```python
gst_prices = list(map(lambda x: x * 2, order_amounts))
gst_prices
```

## Function + Lambda + Array

```python
orders = [30000, 90000, 150000]

def is_premium(amount):
    return amount > 100000

premium_orders = list(filter(lambda x: is_premium(x), orders))

premium_orders
```
