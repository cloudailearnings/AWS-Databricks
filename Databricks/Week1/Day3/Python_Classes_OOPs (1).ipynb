# Python – Classes & OOPs (Real-Time Business Examples)

## 1. Classes and Objects – Employee Management

```python

class Employee:
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def display(self):
        return f"ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary}"

emp1 = Employee(101, "Ravi", 50000)
emp2 = Employee(102, "Anita", 60000)

print(emp1.display())
print(emp2.display())
```

## 2. OOP Concepts – Banking System (Encapsulation & Abstraction)

```python

class BankAccount:
    def __init__(self, account_no, balance):
        self.account_no = account_no
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

acc = BankAccount(12345, 10000)
acc.deposit(-5000)
acc.withdraw(-3000)
print("Current Balance:", acc.get_balance())
```

## 3. Inheritance – Orders & Discounts

```python

class Order:
    def __init__(self, order_id, amount):
        self.order_id = order_id
        self.amount = amount

    def total_amount(self):
        return self.amount

class DiscountOrder(Order):
    def __init__(self, order_id, amount, discount):
        super().__init__(order_id, amount)
        self.discount = discount

    def total_amount(self):
        return self.amount - self.discount

order1 = Order(1, 2000)
order2 = DiscountOrder(2, 3000, 300)

print("Normal Order Total:", order1.total_amount())
print("Discounted Order Total:", order2.total_amount())
```

## 5. Scope – Payroll Calculation

```python

bonus = 2000

def calculate_salary(base_salary):
    bonus = 1000
    return base_salary + bonus

print("Final Salary:", calculate_salary(40000))
print("Global Bonus:", bonus)
```
