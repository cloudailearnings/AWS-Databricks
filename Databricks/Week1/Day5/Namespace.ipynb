# A namespace is a container that holds names (variables, functions, classes) and maps them to objects.

# Think of it like a dictionary:  {name: object}

# Python uses namespaces to avoid name conflicts.

```python
# Without namespaces: Which x is this?

x = 10

def func():
    x = 20
```

# with namespaces: 4 types 

# Built-in Namespace

# Global Namespace

# Local Namespace

# Enclosing Namespace (for nested functions)

# Built-in Namespace - Contains built-in functions and keywords.

```python
import builtins
print(dir(builtins))
```

# Global Namespace - Created when a Python file (module) is loaded.

```python
x = 100

def show():
    print(x)  # x belongs to global namespace

show()
```

# Local Namespace - Created inside a function.

```python
def show():
    y = 50
    print(y) # y exists only inside the function

show()
```

# Enclosing Namespace - Exists for nested functions.

# An enclosing namespace exists when you have a function inside another function.

# Variables defined in the outer function

# Are not global

# But are accessible to inner (nested) functions

# This namespace sits between Local and Global in the LEGB rule.

```python
def outer():
    a = 10

    def inner():
        print(a)

    inner()

outer()
```

# LEGB Rule

"""

Python searches names in this order:

L → E → G → B

     Local

     Enclosing

     Global

     Built-in

"""

```python
x = 5

def outer():
    x = 10
    def inner():
        x = 15
        print(x)
    inner()

outer()
```

# Modifying Names in Namespaces - global keyword




```python
# Used to modify global variables inside a function.

x = 10

def update():
    global x
    x = 20

update()
print(x)
```
