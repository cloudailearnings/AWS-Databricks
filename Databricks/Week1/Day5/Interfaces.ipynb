# An interface defines what a class should do, not how it does it.

# It’s a contract:

# If a class implements an interface

# It must provide implementations for the defined methods

# Python doesn’t have interfaces in the same way Java

# Python supports interfaces using Abstract Base Classes (ABCs) 

# Interfaces using Abstract Base Classes (ABC)

```python
# Payment Interface

from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass
```

```python
# Implementing the Interface

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class UPIPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI")
```

```python
payment = CreditCardPayment()
payment.pay(1000)
```

```python
payment1 = UPIPayment()
payment1.pay(1000)
```
