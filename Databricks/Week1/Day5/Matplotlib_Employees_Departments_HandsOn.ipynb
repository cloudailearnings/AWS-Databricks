# Matplotlib – What, Why & Hands-on Exercises

This notebook explains **what Matplotlib is**, **why it is used**, and provides **hands-on exercises with all major chart types** using Employees & Departments data.

## What is Matplotlib?
Matplotlib is a Python visualization library used to create static, animated, and interactive plots.

## Why Matplotlib?
- Easy data visualization
- Supports many chart types
- Widely used in data analysis and reporting
- Works well with NumPy & Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

employees = pd.read_csv('employees.csv')
departments = pd.read_csv('departments.csv')

employees.head()
#departments.head()
```

## 1. Line Chart – Salary Trend

```python
plt.figure()
plt.plot(employees['EmpID'], employees['Salary'])
plt.xlabel('Employee ID')
plt.ylabel('Salary')
plt.title('Salary Trend')
plt.show()
```

## 2. Bar Chart – Average Salary by Department

```python
avg_salary = employees.groupby('Department')['Salary'].mean()

plt.figure()
plt.bar(avg_salary.index, avg_salary.values)
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salary by Department')
plt.show()
```

## 3. Pie Chart – Employee Distribution

```python
dept_count = employees['Department'].value_counts()

plt.figure()
plt.pie(dept_count.values, labels=dept_count.index, autopct='%1.1f%%')
plt.title('Employee Distribution by Department')
plt.show()
```

## 4. Scatter Plot – Experience vs Salary

```python
plt.figure()
plt.scatter(employees['Experience'], employees['Salary'])
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

## 5. Histogram – Salary Distribution

```python
plt.figure()
plt.hist(employees['Salary'], bins=3)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

## 6. Box Plot – Salary by Department

```python
plt.figure()
employees.boxplot(column='Salary', by='Department')
plt.title('Salary by Department')
plt.suptitle('')
plt.show()
```
