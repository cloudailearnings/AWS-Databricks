## What is Pandas?
Pandas is a powerful Python library used for data manipulation and analysis. It provides fast, flexible data structures like **Series** and **DataFrame**.

## Why Pandas?
- Easy CSV/Excel handling
- Data cleaning and transformation
- Powerful filtering, grouping, aggregation
- Widely used in Data Analytics, Data Science, ML

```python
import pandas as pd

customers = pd.read_csv('customers.csv')
sales = pd.read_csv('sales.csv')

customers, sales
```

##  Basic Data Exploration
- Display first 3 rows
- Check shape and column names


```python
#customers.head(3)

#customers.shape

customers.columns
```

##  Filtering Customers
Find customers from Hyderabad.

```python
customers[customers['age'] == 'Hyderabad']
```

##  High-Value Sales
Filter orders where Amount > 20000.

```python
sales[sales['Amount'] > 20000]
```

## GroupBy Analysis
Calculate total sales per customer.

```python
sales.groupby('Product')['Amount'].sum()
```

## Merge Customers & Sales
Join customer details with sales data.

```python
merged_df = pd.merge(customers, sales, on='CustomerID')
merged_df
```

##  Sales Summary
- Total sales
- Average order value

```python
total_sales = sales['Amount'].sum()
avg_order = sales['Amount'].mean()

total_sales, avg_order
```
