## 9. Hands-on Exercises

1. Create an array of 10 product prices and apply 15% discount
2. Filter orders greater than 1000
3. Create 3x3 matrix and calculate row-wise & column-wise sum


```python
#Create an array of 10 product prices and apply 15% discount

import numpy as np

prices = np.array([500, 750, 1200, 1500, 2000, 2500, 3000, 3500, 4000, 4500])
discounted_prices = prices * 0.85   # 15% discount

discounted_prices
```

```python
#Filter orders greater than 1000

orders = np.array([300, 800, 1200, 450, 1800, 2200, 999, 1500])

high_value_orders = orders[orders > 1000]

high_value_orders
```

```python
# Create 3×3 matrix & calculate row-wise and column-wise sum

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

row_sum = matrix.sum(axis=1)   # Row-wise sum
col_sum = matrix.sum(axis=0)   # Column-wise sum

row_sum, col_sum
```
