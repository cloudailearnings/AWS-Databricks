# Order Total Calculation

```python
import unittest

# Business logic
def calculate_total(price, quantity):
    if price < 0 or quantity < 0:
        raise ValueError("Invalid input")
    return price * quantity


# Test case
class TestMath(unittest.TestCase):

    def test_valid_order(self):
        self.assertEqual(calculate_total(100, 2), 2000)

    def test_invalid_price(self):
        with self.assertRaises(ValueError):
            calculate_total(-100, 2)


# ✅ Jupyter-safe test runner
suite = unittest.TestLoader().loadTestsFromTestCase(TestMath)
unittest.TextTestRunner(verbosity=2).run(suite)

# unittest.TestLoader().loadTestsFromTestCase(...) -- collects all test methods from the class. Finds methods starting with test_ & Converts them into a test suite
# unittest.TextTestRunner() -- Executes the test suite & Prints results to the console (terminal / notebook output)
# By default, it prints: Test names , Pass / Fail , Tracebacks for failures
```
