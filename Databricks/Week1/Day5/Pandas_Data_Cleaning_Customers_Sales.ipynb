# Pandas Data Cleaning & Transformations – Customers & Sales

This notebook focuses on **data cleaning and transformation techniques** using Pandas with real-time **Customers & Sales CSV data**.

## Load Data

```python
import pandas as pd

customers = pd.read_csv('customersinfo.csv')
sales = pd.read_csv('salesinfo.csv')

customers, sales
```

## Data Cleaning – Missing Values
Check and handle missing values.

```python
customers.isnull().sum()

#sales.isnull().sum()
```

```python
# Fill missing ages with average age
customers['Age'] = customers['Age'].fillna(customers['Age'].mean())
customers
```

## Remove Duplicates
Identify and remove duplicate records.

```python
customers
```

```python
customers.duplicated().sum()
```

```python
customers = customers.drop_duplicates()
customers
```

```python
customers.duplicated().sum()
```

## Data Type Corrections
Ensure correct data types.

```python
sales.dtypes

sales['Amount'] = sales['Amount'].astype(float)
```

## Column Transformations
Create new calculated columns.

```python
# Add Discounted Amount (10%)
sales['DiscountedAmount'] = sales['Amount'] * 0.90
sales
```

## Rename & Reorder Columns

```python
#sales = sales.rename(columns={'Amount': 'OriginalAmount'})
sales = sales[['OrderID', 'Product','CustomerID','OriginalAmount','DiscountedAmount']]
sales
```

## Merge Cleaned Data

```python
final_df = pd.merge(customers, sales, on='CustomerID')
final_df
```

## Outlier Detection
Identify high-value sales outliers.

```python
final_df=final_df[final_df['OriginalAmount'] > 50000]
```

## Export Cleaned Data

```python
final_df.to_csv('cleaned_customers_sales_1.csv',index=False)
final_df.head()
```
