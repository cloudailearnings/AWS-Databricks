# Python – Exception Handling


## Error vs Exception

Error :

stops program execution.
An error stops the program immediately. 
You cannot handle it at runtime.

Exception:

An exception happens during execution.

Can handle it using try-except.

Without handling → program crashes
With handling → program continues gracefully

```python
print(10 / 0)
```

# Banking Application (Balance Withdrawal)

```python

# Example of Exception
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Division by zero error")
```

```python
# Without exception handling

balance = 5000
withdraw = int(input("Enter amount: "))

print(balance / withdraw)

print("Thank you for using our bank")
```

```python
# With try-except

balance = 5000

try:
    withdraw = int(input("Enter amount: "))
    result = balance / withdraw
    print("Transaction successful")
except ZeroDivisionError:
    print("Withdrawal amount cannot be zero")
except ValueError:
    print("Please enter a valid number")

print("Thank you for using our bank")
```

```python
# Employee Salary Calculation

try:
    salary = int(input("Enter salary: "))
    bonus = salary * 0.10
    print("Bonus:", bonus)
except ValueError:
    pass

print("Thank you for using our bank")
```

# Division (Multiple except)

```python
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    print(a / b)
except ZeroDivisionError:
    print("❌ Cannot divide by zero")
except ValueError:
    print("❌ Please enter valid numbers")
```

# Dictionary Key Error

```python
order = {"id": 101, "Product": "Laptop"}

try:
    print(order["product"])
except KeyError:
    print("❌ Price key not found")


print("enter a valid key")
```

# List Index Error

```python
products = ["Laptop", "Mobile"]

try:
    print(products[-4])
except IndexError:
    print("❌ Index out of range")

print("privide valid element index")
```

## try-except-else-finally

```python

try:
    num = int("10")
except ValueError:
    print("Value error")
else:
    print("Conversion successful")
finally:
    print("Execution completed")
```

# Custom Exception

```python
class InsufficientBalance(Exception):
    pass

balance = 5000
withdraw = 7000

try:
    if withdraw > balance:
        raise InsufficientBalance("Not enough balance")
except InsufficientBalance as e:
    print("❌", e)
```

```python
class InsufficientBalance(Exception):
    def __init__(self, balance, withdraw):
        self.balance = balance
        self.withdraw = withdraw
        super().__init__(
            f"Withdrawal {withdraw} exceeds balance {balance}"
        )

def withdraw_money(balance, amount):
    if amount < balance:
        raise InsufficientBalance(balance, amount)
    return balance - amount

try:
    balance = 5000
    new_balance = withdraw_money(balance, 7000)
    print("✅ Withdrawal successful. Balance:", new_balance)
except InsufficientBalance as e:
    print("❌ Transaction failed:", e)
```
