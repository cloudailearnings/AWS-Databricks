# Python – File Handling


## Write / Create File

```python

with open("data1.txt", "w") as f:
    f.write("Hello File Handling")
```

```python
# Save employee detail

file = open("employees123.txt", "w")
file.write("EmpId: 101, Name: Rajesh, Dept: IT\n")
file.write("EmpId: 102, Name: Kumar, Dept: HR\n")
file.close()
```

```python
file = open("employees.txt", "a")
file.write("EmpId: 103, Name: Anitha, Dept: Finance\n")
file.close()
```

## Read File

```python
file = open("employees.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
file = open("employees.txt", "r")
for line in file:
    print(line.strip())
file.close()
```

```python

with open("data1.txt", "r") as f:
    print(f.read())
```

## Append File

```python

with open("data1.txt", "a") as f:
    f.write("\nNew Line Added")
```

## Delete File

```python

import os
if os.path.exists("data1.txt"):
    os.remove("data1.txt")
```

```python
import os

if os.path.exists("employees.txt"):
    os.remove("employees1.txt")
    print("File deleted")
else:
    print("File not found")
```
