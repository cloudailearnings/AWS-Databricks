## 2. Connect to MySQL

```python

import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql",
    database="hr"
)

print("Connected:", conn.is_connected())
```

## 3. Create Table

```python

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INT AUTO_INCREMENT PRIMARY KEY,
    emp_name VARCHAR(50),
    salary INT
)
""")

conn.commit()
print("Table ready")
```

## 4. Insert Data

```python

cursor.execute(
    "INSERT INTO employees (emp_name, salary) VALUES (%s, %s)",
    ("Kumar", 50000)
)
conn.commit()
print("Inserted")
```

## 5. Read Data

```python

cursor.execute("SELECT * FROM employees")
for row in cursor.fetchall():
    print(row)
```

## 6. Update Data

```python

cursor.execute(
    "UPDATE employees SET salary=%s WHERE emp_name=%s",
    (55000, "Kumar")
)
conn.commit()
print("Updated")
```

## 7. Delete Data

```python

cursor.execute("DELETE FROM employees WHERE emp_name=%s", ("Kumar",))
conn.commit()
print("Deleted")
```

## 8. Close Connection

```python

cursor.close()
conn.close()
print("Connection closed")
```
